#**1. Data Analysis**


```python
def generate_output(input,device,model,tokenizer):

  model.to(device)


  input=tokenizer(input, return_tensors="pt").to(model.device)
  outputs = model.generate(**input, max_new_tokens=1024)
  return tokenizer.decode(outputs[0], skip_special_tokens=True)[len(input)+1:]
```

```python
from tqdm import tqdm

def output_to_excel(df,model,device,tokenizer):
  outputs={}
  for j in tqdm(df.columns):
      for i in range(len(df[j])):
        if j in outputs.keys():
          if df[j].iloc[i]=='x':
            outputs[j].append(np.nan)
          else:
            outputs[j].append(generate_output(df[j][i],device,model,tokenizer))
        else:
          if df[j].iloc[i]=='x':
            outputs[j]=[np.nan]
          else:
            outputs[j]=[generate_output(df[j][i],device,model,tokenizer)]
  return outputs
```
